**Data Preprocessing**

```python
# drop highly correlated columns
data=data.drop(columns=['maxtemp', 'temparature', 'mintemp',],axis=1)
```

```python
data.head()
```

```python
data['rainfall'].value_counts()
```

```python
# seperate majority and minority class
majority=data[data['rainfall']==1]
minority=data[data['rainfall']==0]
```

```python
print(majority.shape)
```

```python
print(minority.shape)
```

```python
# downssample majority class to match minority count
df_majority_downsample=resample(majority,replace=False,n_samples=len(minority),random_state=42)
data=pd.concat([df_majority_downsample,minority])
```

```python
df_majority_downsample.shape
```

```python
data.head()
```

```python
# shuffle the final datafram
data=data.sample(frac=1,random_state=42).reset_index(drop=True)
```

```python
print(data['rainfall'].value_counts())
```

```python
# splt the features and target as x and y
x=data.drop(columns='rainfall',axis=1)
y=data['rainfall']
```

```python
print(x)
```

```python
# spliting the dataset into training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x.shape,x_test.shape,x_train.shape)
```

```python
rf_model=RandomForestClassifier(random_state=42)
```

```python
prg_grid={
    'n_estimators':[50,100,200],
    'max_features':['sqrt','log2'],
    'max_depth':[None,10,20,30],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]


}
```

```python
# hyperparameter using GridsearchCV
grid_search_cv=GridSearchCV(estimator=rf_model,param_grid=prg_grid,cv=5,n_jobs=-1,verbose=2)
```

```python
grid_search_cv.fit(x_train,y_train)
```

```python
best_rf_model=grid_search_cv.best_params_
print(best_rf_model)
```

```python
best_rf_score=grid_search_cv.best_score_
print(best_rf_score)
```

**Model Evaluation**

```python
cross_val_scores=cross_val_score(rf_model,x_train,y_train,cv=5)
print('Cross val Score of the',rf_model,' = ',np.mean(cross_val_scores)*100)
```

```python
x_train_pred=grid_search_cv.predict(x_train)

print('Accuracy score of the model is=',accuracy_score(y_train,x_train_pred))
print('----------------------------------------------------------------------')
print('confusion Matrix is =',confusion_matrix(y_train,x_train_pred))
print('---------------------------------------------------------------------')
print('classification of the model =',classification_report(y_train,x_train_pred))
```

**Predictive System**

```python
input=(1016.5,16.4,90,90,2.1,40,23.3

)
input_asarray=np.asarray(input)
input_reshape=input_asarray.reshape(1,-1)
prediction=grid_search_cv.predict(input_reshape)
print(prediction)
if prediction[0]==1:
  print('Rainfall')
else:
  print('No Rainfall')
```
